##Importações

```
# Importações necessárias:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# Banco de dados "Video Game Sales and Ratings"
df = pd.read_csv("https://raw.githubusercontent.com/guilhermechicarolli/VideoGames/main/Data%20Base.csv", sep= ";")
df
```

##Análise de popularidade das plataformas dos jogos

```
# Popularidade das plataformas de acordo com as vendas globais (1976-2017)

# Agrupamento da soma das vendas globais em relação às plataformas:
platsales1 = df.groupby('Platform')['Global_Sales'].sum()

# Ordenamento decrescente da soma das vendas 
platsales2 = platsales1.sort_values(ascending = False)

print('Popularidade das plataformas de acordo com as vendas globais (1976-2017):')
platsales2
```

```
# Gráfico da tabela "Popularidade das plataformas de acordo com as vendas globais (1976-2017)"
platforms = ['PS2', 'X360', 'PS3', 'Wii', 'DS', 'PS', 'PS4', 'GBA', 'Outras']
sales = [1251.63, 972.54, 941.22, 908.21, 806.03, 729.70, 340.79, 313.84, 2701]

fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(aspect='equal'))
plt.axis('equal')
plt.pie(sales, labels= platforms, autopct='%1.2f%%')
ax.legend(sales, title='Vendas', loc='right', bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Popularidade das plataformas dos jogos de acordo com as vendas globais (1976-2017):', size = 15, weight='bold')
plt.show()
```

##Os jogos mais bem avaliados pelos usuários e pela crítica

```
# Classificaçao dos 15 jogos mais populares de acordo com avaliação dos usuários

avaliacoes = df[['Name', 'User_Score']].dropna(subset=['User_Score']).sort_values(by=['User_Score'], ascending=[False])
top15 = avaliacoes.head(15)

#Tabela
print('Jogos mais bem avaliados pelos usuários')
top15
```

```
# Classificação dos 15 jogos mais bem avaliados pela crítica

avaliacoes = df[['Name', 'Critic_Score']].dropna(subset=['Critic_Score']).sort_values(by=['Critic_Score', 'Name'], ascending=[False, True])
top15 = avaliacoes.head(15)

#Tabela
print('Jogos mais bem avaliados pela crítica')
top15.drop_duplicates()
```

##Os anos com mais lançamentos de jogos

```
# Classificação dos 10 anos com maiores quantidades de jogos lançados, em porcentagem

anoslanc2 = df['Year_of_Release'].value_counts(ascending=False)[0:7]
outros2 = sum(df['Year_of_Release'].value_counts(ascending=False)[7:])                
todos = df['Year_of_Release'].count()


anoslanc = df['Year_of_Release'].value_counts(ascending=False, normalize=True)[0:7]  #A porcentagem dos 7 anos os quais mais foram lançados jogos
outros = df['Year_of_Release'].value_counts(ascending=False, normalize=True)[7:]    
outros = sum(outros)                                                                  #Porcentagem do resto dos anos

anoslanc2
```

```
# Criação do gráfico de anos com mais lançamentos

valores = [anoslanc[2009], anoslanc[2008], anoslanc[2010], anoslanc[2007], anoslanc[2011], anoslanc[2006], anoslanc[2005], outros]  #Valores do gráfico 
anos = ['2009', '2008', '2010', '2007', '2011', '2006', '2005', 'Outros']

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(aspect='equal'))                              # Tamanho do gráfico
ax.pie(valores, labels = anos, autopct=lambda p: '{:.1f}%\n({:.0f})'.format(p,(p/100)*todos))         #plot do gráfico

ax.legend(anos, title='Anos', loc='right', bbox_to_anchor=(1, 0, 0.5, 1))                            #legenda externa

ax.set_title('Anos com com mais lançamentos de jogos (1976-2017):', size = 15, weight='bold')        #título

plt.show()
```
